## Caricamento librerie/moduli

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import metcomputlib as mc
import myownlib as my
```

```python
mc.Ver()
```

## Caricamento Dati

```python
# load csv in a DataFrame
Aapl = mc.ReadStockPrices('AAPL.csv')

# define the time period of interest
day1 = np.datetime64('2000-01-01')
day2 = np.datetime64('2000-02-28')
```

```python
# extract single columns from DataFrame
D = Aapl['Date'].values
O = Aapl['Open'].values
H = Aapl['High'].values
L = Aapl['Low'].values
C = Aapl['Close'].values
V = Aapl['Volume'].values
```

## Calcolo Medie Mobili con function da libreria myownlib.py

```python
p=7
alpha=0.7
```

```python
SMA=my.Sma(C,p)
WMA=my.Wma(C,p)
EMA=my.Ema(C,p)
EMAW=my.EmaW(C,p)
EMAG=my.EmaG(C,p,alpha)
```

## Figura con function da libreria myownlib.py

```python
plt.figure(figsize=(6, 8), layout='constrained')
# First subplot
plt.subplot(2, 1, 1)
my.PlotData(D, C, day1, day2, 'Close')
my.PlotData(D, SMA, day1, day2, 'SMA')
my.PlotData(D, WMA, day1, day2, 'WMA')
my.PlotData(D, EMA, day1, day2, 'EMA')
mc.EndChart(Legend='on', Xlabel='Days', Ylabel='Prices', Title='Prices and Moving Averages')
# Second subplot
plt.subplot(2, 1, 2)
my.PlotData(D, C, day1, day2, 'Close')
my.PlotData(D, EMA, day1, day2, 'EMA')
my.PlotData(D, EMAW, day1, day2, 'EMAW')
my.PlotData(D, EMAG, day1, day2, 'EMAG')
mc.EndChart(Legend='on', Xlabel='Days', Ylabel='Prices', Title='Exponential Moving Average')
#
plt.show()
```
